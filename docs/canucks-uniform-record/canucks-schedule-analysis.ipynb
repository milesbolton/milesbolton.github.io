```python
import pandas as pd
```

```python
pd.set_option('display.max_rows', 100)
```

```python
df = pd.read_csv('canucks-2324-schedule.csv')
df.head(100)
```

```python
df = df.rename(columns={'Unnamed: 2': 'location'})
```

```python
df['location'] = df['location'].astype(str)
```

```python
df['location'] = df['location'].replace(['nan'],'vs')
df
```

```python
df['Unnamed: 7'] = df['Unnamed: 7'].astype(str)
```

```python
df['Unnamed: 7'] = df['Unnamed: 7'].replace(['nan'],'')
df
```

```python
df['result'] = df['Unnamed: 7'] + df['Unnamed: 6']
df
```

```python
df = df[['Date', 'location', 'Opponent', 'GF', 'GA', 'result', 'Jersey']]
df
```

```python
df['result'] = df['result'].replace(['OTL', 'SOL'], 'OT')
```

```python
df['result'] = df['result'].replace(['OTW', 'SOW'], 'W')
df
```

```python
df.to_csv('schedule-clean.csv')
```

```python
home_df = df.loc[df['Jersey'] == 'Home']
home_df
```

```python
home_record = home_df['result'].value_counts()
print(home_record)
```

```python
home_wins = 15
home_losses = 6
home_ot = 4
```

```python
home_goalsfor = home_df.loc[:, 'GF'].mean().round(2)
home_goalsagainst = home_df.loc[:, 'GA'].mean().round(2)

print('The Canucks scored an average of', home_goalsfor, 'goals, and gave up an average of', home_goalsagainst, 'goals while wearing their home uniform.')
```

```python
road_df = df.loc[df['Jersey'] == 'Road']
road_df
```

```python
road_record = road_df['result'].value_counts()
print(road_record)
```

```python
road_wins = 23
road_losses = 14
road_ot = 4
```

```python
road_goalsfor = road_df.loc[:, 'GF'].mean().round(2)
road_goalsagainst = road_df.loc[:, 'GA'].mean().round(2)

print('The Canucks scored an average of', road_goalsfor, 'goals, and gave up an average of', road_goalsagainst, 'goals while wearing their road uniform.')
```

```python
skate_df = df.loc[df['Jersey'] == 'Skate']
skate_df
```

```python
skate_record = skate_df['result'].value_counts()
print(skate_record)
```

```python
skate_wins = 12
skate_losses = 4
skate_ot = 1
```

```python
skate_goalsfor = skate_df.loc[:, 'GF'].mean().round(2)
skate_goalsagainst = skate_df.loc[:, 'GA'].mean().round(2)

print('The Canucks scored an average of', skate_goalsfor, 'goals, and gave up an average of', skate_goalsagainst, 'goals while wearing their skate uniform.')
```

```python
columns = ['Uniform', 'Wins', 'Losses', 'OT', 'GF', 'GA']
data = [('Home', home_wins, home_losses, home_ot, home_goalsfor, home_goalsagainst), ('Road', road_wins, road_losses, road_ot, road_goalsfor, road_goalsagainst), ('Skate', skate_wins, skate_losses, skate_ot, skate_goalsfor, skate_goalsagainst)]
jersey_record = pd.DataFrame(data, columns=columns)
jersey_record
```

```python
jersey_record.to_csv('uniform-statistics.csv')
```
