```python
import requests
import pandas as pd
```

```python
df = pd.read_csv("base-data.csv")
df.head(100)
```

```python
avg1994 = df["cap1994"].mean().round()
avg1994
```

```python
avg2004 = df["cap2004"].mean().round()
avg2004
```

```python
avg2014 = df["cap2014"].mean().round()
avg2014
```

```python
avg2024 = df["cap2024"].mean().round()
avg2024
```

```python
columns = ['Year', 'avgCap']
data = [('1994', avg1994), ('2004', avg2004), ('2014', avg2014), ('2024', avg2024)]
overallchange_df = pd.DataFrame(data, columns=columns)
overallchange_df
```

```python
overallchange_df.plot.line(x='Year', y='avgCap')
```

```python
df['attend_diff'] = df['cap1994'] - df['cap2024'] 
print("\nDifference of score1 and score2 :\n", df) 
```

```python
df.sort_values('attend_diff', ascending=False).plot.barh(x='Team', y='attend_diff')
# df.sort_values('attend_diff', ascending=False)[['Team','attend_diff']].plot.barh(stacked=True)
```
