```python
import pandas as pd
```

```python
pd.set_option('display.max_rows', 100)
```

```python
df = pd.read_csv('canucks2324jerseys-with-standings.csv')
df.head(100)
```

```python
df = df.rename(columns={'Unnamed: 2': 'location'})
```

```python
df['location'] = df['location'].astype(str)
```

```python
df['location'] = df['location'].replace(['nan'],'vs')
df
```

```python
df['Unnamed: 7'] = df['Unnamed: 7'].astype(str)
```

```python
df['Unnamed: 7'] = df['Unnamed: 7'].replace(['nan'],'')
df
```

```python
df['result'] = df['Unnamed: 7'] + df['Unnamed: 6']
df
```

```python
df = df[['Date', 'location', 'Opponent', 'GF', 'GA', 'result', 'van_rank', 'opp_rank', 'Jersey']]
df
```

```python
df['result'] = df['result'].replace(['OTL', 'SOL'], 'OT')
```

```python
df['result'] = df['result'].replace(['OTW', 'SOW'], 'W')
df
```

```python
df.to_csv('schedule-clean.csv')
```

```python
home_df = df.loc[df['Jersey'] == 'Home']
home_df
```

```python
home_van = home_df.loc[:, 'van_rank'].mean().round()
home_van
```

```python
home_opp = home_df.loc[:, 'opp_rank'].mean().round()
home_opp
```

```python
home_record = home_df['result'].value_counts()
print(home_record)
```

```python
home_wins = 15
home_losses = 6
home_ot = 4
```

```python
home_goalsfor = home_df.loc[:, 'GF'].mean().round(2)
home_goalsagainst = home_df.loc[:, 'GA'].mean().round(2)

print('The Canucks scored an average of', home_goalsfor, 'goals, and gave up an average of', home_goalsagainst, 'goals while wearing their home uniform.')
```

```python
columns = ['Uniform', 'Wins', 'Losses', 'OT', 'GF', 'GA', 'van_rank', 'opp_rank']
data = [('Home', home_wins, home_losses, home_ot, home_goalsfor, home_goalsagainst, home_van, home_opp), ('Road', road_wins, road_losses, road_ot, road_goalsfor, road_goalsagainst, road_van, road_opp), ('Skate', skate_wins, skate_losses, skate_ot, skate_goalsfor, skate_goalsagainst, skate_van, skate_opp)]
jersey_record = pd.DataFrame(data, columns=columns)
jersey_record
```
