```python
df = pd.read_csv('canucks2324jerseys-with-standings.csv')
# df.head(100)
```

```python
df = df.rename(columns={'Unnamed: 2': 'location'})
```

```python
df['location'] = df['location'].astype(str)
```

```python
df['location'] = df['location'].replace(['nan'],'vs')
```

```python
df['Unnamed: 7'] = df['Unnamed: 7'].astype(str)
```

```python
df['Unnamed: 7'] = df['Unnamed: 7'].replace(['nan'],'')
```

```python
df['result'] = df['Unnamed: 7'] + df['Unnamed: 6']
```

```python
df = df[['Date', 'location', 'Opponent', 'GF', 'GA', 'result', 'van_rank', 'opp_rank', 'Jersey']]
```

```python
df['result'] = df['result'].replace(['OTL', 'SOL'], 'OT')
```

```python
df['result'] = df['result'].replace(['OTW', 'SOW'], 'W')
```
