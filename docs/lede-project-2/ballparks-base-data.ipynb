```python
import requests
import pandas as pd
```

```python
df = pd.read_csv("base-data.csv")
df.head(100)
```

```python
avg1994 = df["cap1994"].mean().round()
avg1994
```

```python
avg2004 = df["cap2004"].mean().round()
avg2004
```

```python
avg2014 = df["cap2014"].mean().round()
avg2014
```

```python
avg2024 = df["cap2024"].mean().round()
avg2024
```

```python
columns = ['Year', 'avgCap']
data = [('1994', avg1994), ('2004', avg2004), ('2014', avg2014), ('2024', avg2024)]
overallchange_df = pd.DataFrame(data, columns=columns)
overallchange_df
```

```python
overallchange_df.plot.line(x='Year', y='avgCap')
```

```python
df['cap_diff'] = df['cap1994'] - df['cap2024'] 
df
```

```python
change_df = df.drop(columns=['cap1994', 'cap2004', 'cap2014', 'cap2024'])
change_df
```

```python
change_df = change_df.drop([0, 17, 27, 30])
change_df
```

```python
# change_df(['attend_diff'] * -1)
# df['Revenue_with_*'] = df['Revenue'] * 1.1
change_df['cap_diff'] = change_df['cap_diff'] * -1
change_df
```
