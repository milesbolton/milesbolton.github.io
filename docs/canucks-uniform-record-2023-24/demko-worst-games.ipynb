```python
import pandas as pd
```

```python
df = pd.read_csv('demko-game-log.csv')
df
```

```python
home_demko = df.loc[df['Uniform'] == 'Home']
home_demko
```

```python
home_svpct = home_demko.loc[:, 'SV%'].mean().round(3)
home_svpct
```

```python
home_ga = home_demko.loc[:, 'GA'].mean().round(3)
home_ga
```

```python
road_demko = df.loc[df['Uniform'] == 'Road']
road_demko
```

```python
road_svpct = road_demko.loc[:, 'SV%'].mean().round(3)
road_svpct
```

```python
road_ga = road_demko.loc[:, 'GA'].mean().round(3)
road_ga
```

```python
skate_demko = df.loc[df['Uniform'] == 'Skate']
skate_demko
```

```python
skate_svpct = home_demko.loc[:, 'SV%'].mean().round(3)
skate_svpct
```

```python
skate_ga = skate_demko.loc[:, 'GA'].mean().round(3)
skate_ga
```

```python
columns = ['Uniform', 'SV%', 'GAA']
data = [('Home', home_svpct, home_ga), ('Road', road_svpct, road_ga), ('Skate', skate_svpct, skate_ga)]
demko_record = pd.DataFrame(data, columns=columns)
demko_record
```

```python
demko_record.to_csv('demko-uniform-stats')
```
