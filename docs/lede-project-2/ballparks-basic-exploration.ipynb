```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
ballparks_url = "https://en.wikipedia.org/wiki/List_of_current_Major_League_Baseball_stadiums"
ballparks_html = requests.get(ballparks_url).text
```

```python
ballparks_soup = BeautifulSoup(ballparks_html)
```

```python
tables = ballparks_soup.select("table")
len(tables)
```

```python
tables[0]
```

```python
ballparks_table = tables[0]
ballparks_table.text[:100]
```

```python
row_els = ballparks_table.select("tbody tr")
len(row_els)
```

```python
row_els[0]
```

```python
first_row = row_els[0]
first_row_cells = first_row.select("th")
first_row_cells
```

```python
ballparks_headers = first_row_cells
ballparks_headers
```

```python
for cell in ballparks_headers:
    print(cell.text.strip())
```

```python
ballparks_entries = []

for row in row_els:
    row_cells = []
    for cell in row.select("th"):
        row_cells.append(cell.text.strip())    
    for cell in row.select("td"):
        row_cells.append(cell.text.strip())
    ballparks_entries.append(row_cells)

ballparks_entries[:3]
```

```python
ballparks_df = pd.DataFrame(ballparks_entries, columns=ballparks_headers)
ballparks_df.head()
```
