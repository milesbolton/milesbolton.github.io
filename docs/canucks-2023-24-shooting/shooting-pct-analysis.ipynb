```python
import pandas as pd
```

```python
df = pd.read_csv('jt-miller-game-log.csv')
df.head(100)
```

```python
df.dtypes
```

```python
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
```

```python
df
```

```python
df = df.drop(1)
```

```python
# df = df.rename(columns={'NaN': 'home-away'})
# df
df.columns = df.columns.fillna('home-away')
```

```python
df['home-away'] = df['home-away'].astype(str)
df
```

```python
df['home-away'] = df['home-away'].replace(['nan'],'vs')
df
```

```python
df['game'] = df['Date'] + ' (' + df['Tm'] + ' ' + df['home-away'] + ' ' + df['Opp'] + ')'
df
```

```python
df = df[['game', 'G', 'S', 'S%']]
df
```

```python
pd.set_option('display.max_rows', 100)
```

```python
df
```

```python
df.dtypes
```

```python
# df["S%"] = pd.to_numeric(df["S%"])
# df['S%'] = df['S%'].astype(float)
df["S%"] = pd.to_numeric(df["S%"], errors='coerce')
```

```python
pd.options.mode.chained_assignment = None
```

```python
df.dtypes
```

```python
# change NaN in S% to 0 // ignore NaN values when calcutlating rolling avg
# df.fillna(0)  wrong place
```

```python
# df['MA'] = df.rolling(window=5).mean()
df['rolling-avg'] = df['S%'].rolling(5).mean().round(1)
df
```

```python
df.fillna(0)
```

```python
df['rolling-avg'].plot()
```

```python
df.to_csv('PLAYER-shooting-rolling-avg.csv')
```
